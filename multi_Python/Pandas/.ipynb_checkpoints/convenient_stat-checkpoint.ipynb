```python
##7. 영등포구에서 시간 당 급여가 6000원 이상인 편의점 검색
con[(con['area1'] == '영등포구')&(con['hourly_wage'] >= 6000)]
```

```python
##8. CU 편이점만 출력 (상위 10개만)
con[con['name'] == 'CU'].head(10)
```

```python
##9. 지역 컬럼(location)을 추가한 다음, in Seoul 이라는 값 저장, 상위 5개 출력
con['location'] = 'in Seoul'
con.head(5)
```

```python
##10. 6000원 이상 컬럼 추가(more_than_6000) -> True, False 값 저장 (상위 20개 출력)
def check(wage):
    if wage >= 6000:
        return True
    else:
        return False
con['more_than_6000'] = con['hourly_wage'].apply(lambda x : check(x))
con.head(20)
```

```python
##11. more_than_6000 컬럼에서 True인 데이터들의 평균, 개수, 편차 등의 정보 출력
```

```python
# 평균
con[con['more_than_6000'] == True]['hourly_wage'].mean()
```

```python
# 개수
con[con['more_than_6000'] == True]['hourly_wage'].count()
```

```python
# 편차
con[con['more_than_6000'] == True]['hourly_wage'].std()
```

```python
##12. more_than_6000 이름의 함수를 생성하고, 6000원이상인 경우 A group, 아니면 B group을 반환하는 함수 생성
def more_than_6000(wage):
    if wage>= 6000:
        return 'A'
    else:
        return 'B'
```

```python
##13. more_than_6000_f 컬럼 생성하고 more_than_6000 함수의 결과를 저장
con['more_than_6000_f'] = con['hourly_wage'].apply(lambda x : more_than_6000(x))
```

```python
##14. 지금까지의 결과 상위 10개를 출력
con.head(10)
```

```python
##15-1. more_than_6000가 True인 데이터의 지역과 시간당 급여를 가진 새로운 데이터프레임 생성(data2)
##15-2. data2 데이터를 시간당 급여 순으로 정렬 (높은순)
data2 = con[con['more_than_6000'] == True][['area', 'hourly_wage']]
data2 = data2.sort_values(by='hourly_wage', axis=0, ascending=False)
```

```python
##16. data2를 darta2.csv 파일로 저장
data2.to_csv('darta2.csv', index=False)
```
